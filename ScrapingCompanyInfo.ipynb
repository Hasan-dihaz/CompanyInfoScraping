```python
from selenium import webdriver  #automation proceess library
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from time import sleep
import csv
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
#from goto import goto,label

options = webdriver.ChromeOptions()
ser = Service("C:\chromedriver_win32\chromedriver.exe")
driver = webdriver.Chrome(service=ser, options=options)
driver.maximize_window() # to open full size window
#driver = webdriver.Chrome("C:\chromedriver_win32\chromedriver.exe") #Installtion path or chrome web driver 
driver.get("https://www.google.com/maps/@31.4855012,74.3470166,15z")#https://www.google.com/maps/@1.3011523,103.8510006,17z
sleep(5) 

#xpath is stand for XML path language, uses path expression to identify and navigate nodes in an XML document can be use to 
#select  one/more nodes in the document by using absolute / relative path 

def searchplace():
    Place = driver.find_element(By.CLASS_NAME, "tactile-searchbox-input")
    Place.send_keys("It Solutions, software company in singapore")# Singapore software company(started in 6th search)...Software company in Singapore
    submit = driver.find_element(By.XPATH, "/html/body/div[3]/div[9]/div[3]/div[1]/div[1]/div[1]/div[2]/div[1]/button")
    submit.click()
searchplace()
```

```python
element = driver.find_element(By.CLASS_NAME, "w6VYqd")
```

```python
elements=element.find_elements(By.CLASS_NAME, "hfpxzc")#......Reads only visible names in home page......
```

```python
from selenium.webdriver.common.action_chains import ActionChains
actions = ActionChains(driver)

count=0
m_lis=[]
for i in elements:
    count+=1
    if i.is_displayed():
        #i.click()
        driver.execute_script("arguments[0].click();", i)
    else:
        actions.move_to_element(i).perform()#.......scrolling
        sleep(2)
        i.click()
    sleep(12)
    
    lis=list() 
    name = driver.find_element(By.CLASS_NAME,"DUwDvf")
    lis.append(name.text)
    print(name.text)
    
#..................................///////////////////................................Reading details per company....
    
    contact_no=str('no_contact')
    plus_code=str('no_plus_codes')
    web=str('no_web')
    
    addresses=driver.find_elements(By.CLASS_NAME,"RcCsl")
    for add in addresses:#------------------------------------------------|||
        txt = add.find_element(By.CLASS_NAME,"Io6YTe").text
#         print('txt: ', txt)
#         print('len: ',len(lis))       
        if len(lis)<=2:
            lis.append(txt)
        if '+' in txt:
            indx = txt.index('+')
            if indx==0:#.............................+65 6220 5658
                contact_no = txt
                #print("con: ",contact_no)
            else:#...................................8V23+F7 Singapore
                plus_code = txt
                #print("plus: ", plus_code)
                
        if '.' in  txt:#............................alphaxlab.co
            #print(".: ",txt)
            web = txt#-----------------------------------------------------|||
            
#..............................///////////////.................................                     
            
#     match = {contact_no, plus_code, web}#.......works too..........
#     if lis[2] in match:
#         lis[2] = ' '
    if lis[2].startswith((contact_no, plus_code, web)):#................overwriting list value at position 2
        lis[2] = ' '
        
# .........................../////////////////.......................Adding all data in the list per company
        
    if len(lis)>2:
        lis.append(web)
        lis.append(contact_no)
        lis.append(plus_code)
#     else:
#         lis.append(' ')
        
#     print("con: ",contact_no)
#     print("plus: ",plus_code)
#     print("web: ", web)
#     print('\n\n')
        
# ............................//////////...................................Ends of Reading details per company....


    rating_review = driver.find_element(By.CSS_SELECTOR, 'div.F7nice')#.......reading rating & review
# review = driver.find_element(By.CSS_SELECTOR, 'span.DkEaL')
    txt_rating_review = rating_review.text
    rat_rev_word_list = txt_rating_review.split('\n')
    rating = rat_rev_word_list[0]#.............................rating.    
    lis.append(rating)#.......................................adding rating to the list
        
    if len(rat_rev_word_list)>1:#...............................reading review
        txt_review = rat_rev_word_list[1]
        rev_word_list = txt_review.split(" ")
        review_number = rev_word_list[0]#.......................review
        lis.append(review_number)#..............................adding review to the list
        
    
    m_lis.append(lis)#................Final list per company
        
#     if count==10:
#         break
```
